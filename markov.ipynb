```python
import numpy as np
```

```python
pi_0 = np.array([0.40, 0.35, 0.20, 0.05])
```

```python
P = np.array([
    [0.85, 0.06, 0.07, 0.02],
    [0.08, 0.80, 0.10, 0.02],
    [0.10, 0.10, 0.75, 0.05],
    [0.12, 0.10, 0.08, 0.70]
])
```

```python
pi_1 = np.dot(pi_0, P)
print("Market shares for period 1:")
print(pi_1)
```

```python
pi_2_from_1 = np.dot(pi_1, P)
print('Market shares for period 2 given period 1:')
print(pi_2_from_1)
```

```python
P2 = np.dot(P, P)
print(P2)
```

```python
pi_2_direct = np.dot(pi_0, P2)
print('Market shares for period 2 given period 0:')
print(pi_2_direct)
```

```python
pi_0 = np.array([1, 0])
```

```python
P = np.array([
    [0.90, 0.10],
    [0.20, 0.80]
])
```

```python
n_periods = 20
```

```python
state_probabilities = [pi_0]
```

```python
for _ in range(n_periods):
    pi_next = np.dot(state_probabilities[-1], P)
    state_probabilities.append(pi_next)
```

```python
for i, pi in enumerate(state_probabilities):
    print(f'Period {i} = {pi}')
```

```python
P = np.array([
    [1, 0, 0, 0],
    [0, 1, 0, 0],
    [0.80, 0, 0.10, 0.10],
    [0.30, 0.10, 0.30, 0.30]
])
```

```python
I = np.array([
    [1, 0],
    [0, 1]
])
```

```python
A = np.array([
    [0.80, 0],
    [0.30, 0.10]
])
```

```python
B = np.array([
    [0.10, 0.10],
    [0.30, 0.30]
])
```

```python
I_minus_B = I - B
F = np.linalg.inv(I_minus_B)

FA = np.dot(F, A)

print('The final matrix FA is:')
print(FA)
```
